Table Patient

```python
fake = Faker('en_US')
Faker.seed(42)
random.seed(42)


num_records = 10000


columns = [
    'patient_id', 'first_name', 'last_name', 'gender', 'age',
    'height_cm', 'weight_kg', 'country', 'city', 'visits_count']


data = []

countries_list = ['Egypt', 'Saudi Arabia', 'UAE', 'Jordan', 'Morocco', 'Kuwait', 'Qatar']

for _ in range(num_records):
    record = {
        'patient_id': fake.unique.random_number(digits=5),
        'first_name': fake.first_name(),
        'last_name': fake.last_name(),
        'gender': random.choice(['Male', 'Female']),
        'age': random.randint(1, 90),
        'height_cm': random.randint(100, 200),
        'weight_kg': random.randint(40, 200),
        'country': random.choice(countries_list),
        'city': fake.city(),
        'visits_count': random.randint(1, 100)
    }
    data.append(record)


df = pd.DataFrame(data, columns=columns)


print(df.head())


df.to_csv('fake_patient_data.csv', index=False)
```

Table Payment

```python
fake = Faker('en_US')
Faker.seed(42)
random.seed(42)

num_records = 10000

columns = ['payment_id','method','amount','payment_date','payment_status','transaction_id']

data = []


start_datetime = datetime(2023, 1, 1)
end_datetime = datetime(2023, 12, 31)

for _ in range(num_records):
    record = {
        'payment_id': fake.unique.random_number(digits=5),
        'method': random.choice(['cash','credit_card','debit_card','insurance','online']),
        'amount': round(random.uniform(10, 5000), 2),
        'payment_date':fake.date_time_between(start_date=start_datetime, end_date=end_datetime),
        'payment_status': random.choice(['pending', 'completed', 'failed', 'refunded']),
        'transaction_id': fake.unique.random_number(digits=5)
    }
    data.append(record)

df = pd.DataFrame(data, columns=columns)

print(df.head())

df.to_csv('fake_payment_data.csv', index=False)
```

Table Department

```python
fake = Faker('en_US')
Faker.seed(42)
random.seed(42)

num_records = 10000

columns = ['department_id','department_name','department_code','current_occupancy','max_capacity','num_staff','working_hours','emergency_support']

data = []

dept_names = [
    'Cardiology', 'Neurology', 'Orthopedics', 'Pediatrics',
    'Radiology', 'Surgery', 'Dermatology', 'Oncology',
    'Emergency', 'ICU', 'Laboratory', 'Pharmacy'
]

for _ in range(num_records):
    record = {
        'department_id': fake.unique.random_number(digits=5),
        'department_name': random.choice(dept_names),
        'department_code': fake.unique.bothify(text='DEP-??##'),
        'current_occupancy': random.randint(0, 200),
        'max_capacity': random.randint(50, 500),
        'num_staff': random.randint(1, 100),
        'working_hours': f"{random.randint(6,12)} AM - {random.randint(4,10)} PM",
        'emergency_support': random.choice(['yes', 'no'])
    }
    data.append(record)

df = pd.DataFrame(data, columns=columns)

print(df.head())

df.to_csv('fake_department_data.csv', index=False)
```

Table Doctor

```python
fake = Faker('en_US')
Faker.seed(42)
random.seed(42)

num_records = 10000

columns = ['doctor_id','first_name','last_name','age','email','gender','specialization','graduation_year','university_grade','educational_degree','hire_year','years_of_experience','rating_avg','salary']

data = []

graduation_year = random.randint(1980, 2025)

for _ in range(num_records):
    record = {
        'doctor_id': fake.unique.random_number(digits=5),
        'first_name': fake.first_name(),
        'last_name': fake.last_name(),
        'age': random.randint(23,100),
        'email': fake.email(),
        'gender': random.choice(['Male', 'Female']),
        'specialization': random.choice(['Cardiology', 'Neurology', 'Orthopedics', 'Pediatrics',
                                         'Radiology', 'Surgery', 'Dermatology', 'Oncology',
                                         'Emergency', 'ICU', 'Laboratory', 'Pharmacy']),
        'graduation_year': random.randint(1980, 2025),
        'university_grade': random.choice(['A', 'B', 'C', 'D', 'E']),
        'educational_degree': random.choice(['Bachelor', 'Master', 'PhD']),
        'hire_year': random.randint( graduation_year + 1, 2025),
        'years_of_experience': random.randint(1, 40),
        'rating_avg': round(random.uniform(1, 5), 2),
        'salary': round(random.uniform(2000, 100000), 2)
    }

    data.append(record)

df = pd.DataFrame(data, columns=columns)

print(df.head())

df.to_csv('fake_doctor_data.csv', index=False)
```
